```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
file_path=r"C:\Users\user\Downloads\Visadataset (1).csv"
Visa_df=pd.read_csv(file_path)
Visa_df
```

```python
keys=Visa_df['continent'].value_counts().keys()
values=Visa_df['continent'].value_counts().values
```

```python
types=Visa_df.dtypes
cat=[i for i,j in types.items() if j=='object']
cat
```

```python
num=[i for i,j in types.items() if j!='object']
num
```

```python
import os
folder_name="Chart_BAR"
try:
    os.makedirs(folder_name)
except Exception as e:
    print(e)
cwd=os.getcwd()
new_dir=os.path.join(cwd,folder_name)

for i in cat[1:]:
    keys=Visa_df[i].value_counts().keys()
    values=Visa_df[i].value_counts().values
    file_name=f"{i}_bar.jpg"
    path=os.path.join(new_dir,file_name)
    plt.bar(keys,values)
    plt.savefig(path)
```
